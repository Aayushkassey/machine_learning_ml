#NumPy Array Manipulation

```python
import numpy as np
```

Transposing Arrays

```python
b = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Original array b:")
print(b)
print("Shape:", b.shape)
```

```python
i = np.transpose(b)
print("\nTranspose using np.transpose():")
print(i)
```

```python
print("\nTranspose using .T:")
print(b.T)
```

Changing Array Shape

Flatten the array

```python
b = np.array([[1, 2, 3],
              [4, 5, 6]])
flat = b.ravel() # flattens data
print("Flattened array:")
print(flat)
print("Shape:", flat.shape)
```

flatten() → copy <br>
ravel() → view

```python
a = np.arange(6)                  # [0, 1, 2, 3, 4, 5]
b = a.reshape((2,3))              # view (if possible)
print("b:\n", b)
print("b.flatten():", b.flatten())# flatten creates a new copy in the memory
print("b.ravel():", b.ravel())    # ravel returns a view, not a copy
```

View vs Copy

```python
a = np.arange(6)
print(a)
v = a.view()      # shallow view (shares memory)
c = a.copy()      # deep copy (independent)

v[0] = 999
print("original after changing view:", a[0])  # changed because view shares memory

c[1] = -1
print("original after changing copy:", a[1])  # unchanged
```

Reshape the array

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
g = a.reshape(3, 2)
print("\nReshaped array:")
print(g)
print("Shape:", g.shape)
```

Adding / Removing Elements
<br><br>
Resizing the array: repeats data if needed

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
resized = np.resize(a, (2, 6))
print("Resized array (2×6):")
print(resized)
```

Append elements

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
appended = np.append(a, [9,8,7])
print("\nAppended array:")
print(appended)
```

Insert Item

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
inserted = np.insert(a, 1, 99)
print("\nAfter inserting 99 at index 1:")
print(inserted)
```

Delete Item

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])
deleted = np.delete(a, [1])
print("\nAfter deleting index 1:")
print(deleted)
```

Combining Arrays:

```python
a = np.array([1, 2, 3])
b = np.array([10, 15, 20])
```

```python
# Concatination
concat = np.concatenate((a, b), axis=0)
print("Concatenated array:")
print(concat)
```

```python
# Vertical Stack
vstacked = np.vstack((a, b))
print("\nVertical stack:")
print(vstacked)
```

```python
e = np.array([[7],
              [7]])

f = np.array([[1],
              [0]])

hstacked = np.hstack((e, f))
print("\nHorizontal stack:")
print(hstacked)
```

Splitting Arrays

```python
x = np.array([1, 2, 3, 4])

# horizontal split
hs = np.hsplit(x, 2)
print("Horizontal split:")
print(hs)

# vertical split
y = np.array([[1, 2, 3],
              [4, 5, 6]])

vs = np.vsplit(y, 2)
print("\nVertical split:")
print(vs)
```

Linear Algebra with NumPy

```python
#Simple Linear Algebra
A = np.array([[3,1],[1,2]], dtype=float)
b = np.array([9,8], dtype=float)

# matrix multiply
x = np.linalg.solve(A, b)   # solves Ax = b -> x = ?
print("solution x:", x)

# inverse & determinant
invA = np.linalg.inv(A)
detA = np.linalg.det(A)
print("invA:\n", invA)
print("det(A):", detA)

# eigen decomposition
vals, vecs = np.linalg.eig(A)
print("eigenvalues:", vals)
```

###Handling Conditional Transformation:

```python
#Masked array concept
a = np.array([-2, -1, 0, 1, 2])

# replace negatives with 0
b = np.where(a < 0, 0, a*10)
print("replace negatives:", b)

# boolean checks
print("any negative?", np.any(a < 0))
print("all non-negative?", np.all(a >= 0))

# masked array (NumPy masked)
import numpy.ma as ma
masked = ma.masked_where(a < 0, a)   # mask negatives
print("masked:", masked)
```

Images as an array

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


img = mpimg.imread('dog.jpg') # Reads the file into a NumPy array.
# print(img)

print(img.shape) # return height, width, channels(3 or 4(RGBA))
height = img.shape[0]
width = img.shape[1]

# invert colors using NumPy
img_inv = 1-img[:,:,:3]  # ignore alpha channel in png
img_rot = np.rot90(img)

# show (in a notebook)
plt.subplot(1,3,1); plt.title("orig"); plt.imshow(img); plt.axis('off')
plt.subplot(1,3,2); plt.title("inverted"); plt.imshow(img_inv); plt.axis('off')
plt.subplot(1,3,3); plt.title("rotated 90"); plt.imshow(img_rot); plt.axis('off')
plt.show()
```
